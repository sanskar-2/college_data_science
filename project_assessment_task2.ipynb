```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('data.xlsx')
data
```

```python
data.info()
```

```python
data = data[data.index < 1000]
data
```

```python
data.info()
```

```python
data = data[['ID','Salary','Designation','JobCity','DOB','Gender','Degree','collegeGPA']]
data.info()
```

```python
data
```

```python
from datetime import date
from datetime import datetime, date

data['DOB'] = data['DOB'].astype('string')

def age(born):
    born = datetime.strptime(born,"%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month,today.day) < (born.month,born.day))

data['Age'] = data['DOB'].apply(age)
data.info()
```

```python
data3 = data
```

```python
data['Gender'].value_counts()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
# ax.title("Total population 1000")
arr2 = np.array(data['Gender'].value_counts())
arr = ['Male','Female']
ax.pie(arr2, labels = arr,autopct='%1.2f%%')
# plt.title(label=)
plt.legend(title ="total population is 1000" )
plt.show()
```

```python
data['Degree'].value_counts()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
arr2 = np.array(data['Degree'].value_counts())
arr = ['B.Tech/B.E.','MCA','M.Tech/M.E.']
ax.pie(arr2, labels = arr,autopct='%1.2f%%',explode=[0,0.2,0.4])
plt.legend(title ="Degree" )
plt.show()
```
